# Task4: Restaurant Chains

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = pd.read_csv('restaurants.csv')
```

### Identify if there are any restaurant chains present in the dataset.

```python
restaurant_chain=dt.groupby("Restaurant Name").size().reset_index(name="OutletCount")
new=restaurant_chain[restaurant_chain["OutletCount"]>1]
restaurant_chains=new.sort_values(by="OutletCount",ascending=False)
print(restaurant_chains[["Restaurant Name","OutletCount"]].head(10))
```

```python
plt.bar(restaurant_chains["Restaurant Name"].iloc[:5],restaurant_chains["OutletCount"].iloc[:5],color='lightgreen')
plt.xlabel("Restaurant Chain")
plt.ylabel("Number of Outlets")
plt.title("Top 5 Restaurant Chains by Outlet Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.figure(figsize=(8,4))
plt.show()
```

### Analyze the ratings and popularity of different restaurant chains.

```python
ratings=dt.groupby("Restaurant Name")["Aggregate rating"].mean().reset_index(name="Average Rating").sort_values(by="Average Rating", ascending=False)
print(ratings)
```

```python
votes=dt.groupby("Restaurant Name")["Votes"].sum().reset_index(name="Total Votes").sort_values(by="Total Votes",ascending=False)
print(votes)
```

```python
plt.bar(ratings["Restaurant Name"][:5],ratings["Average Rating"][:5],color="lightgreen")
plt.xlabel("Restaurant Chain")
plt.ylabel("Average Rating")
plt.title("Top 5 Restaurant Chains by Average Rating")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(votes["Restaurant Name"][:5],votes["Total Votes"][:5],color="lightgreen")
plt.xlabel("Restaurant Chain")
plt.ylabel("Total Votes")
plt.title("Top 5 Restaurant Chains by Total Votes")
plt.tight_layout()
plt.xticks(rotation=90,ha='right')
plt.show()
```
